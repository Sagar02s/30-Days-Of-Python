Iterate 0 to 10 using for loop, do the same using while loop.

```python
for i in range(11):
    print(i)

i=0
while i <=10:
    print(i)
    i += 1
```

Iterate 10 to 0 using for loop, do the same using while loop.

```python
for i in range(10,-1,-1):
    print(i)

i = 10
while i>=0:
    print(i)
    i -= 1
```

Use for loop to iterate from 0 to 100 and print the sum of all evens and the sum of all odds.

The sum of all evens is 2550. And the sum of all odds is 2500.

```python
even_summation = 0
for i in range(101):
    if i%2 ==0:
        even_summation += i
        print(even_summation)
        
odd_summation = 0
for i in range(101):
    if i%2 !=0:
        odd_summation += i
        print(odd_summation)
        
```

Go to the data folder and use the countries_data.py file.
What are the total number of languages in the data
Find the ten most spoken languages from the data
Find the 10 most populated countries in the world

```python
import sys
sys.path.insert(1,'../data')
import countries_data



# What are the total number of languages in the data
languages = []
for country_data in countries_data.countries_data:
    languages.extend(country_data['languages'])
print("The total number of languages are: ",len(set(languages)))

# Find the ten most spoken languages from the data
for language in languages:
    print(language ,":",languages.count(language))


# Find the 10 most populated countries in the world
population = {}
for country_data in countries_data.countries_data:
    population[country_data['name']] = country_data['population']

print((population))
```
